```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from wordcloud import WordCloud, STOPWORDS
import textblob 
from textblob import TextBlob
from subprocess import check_output
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("Reviews.csv")
df
```

```python
df["polarity"] = df["tweets"].apply(lambda x: TextBlob(x).sentiment.polarity)
df["subjectivity"] = df["tweets"].apply(lambda x: TextBlob(x).sentiment.subjectivity)

df
```

```python

stopwords = set(STOPWORDS)
text = " ".join(review for review in df.tweets)
wordcloud = WordCloud(stopwords=stopwords).generate(text)
plt.imshow(wordcloud)
plt.axis("off")
#to save add plt.savefig('wordcloud11.png')
plt.show()
```
